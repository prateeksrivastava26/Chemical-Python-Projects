# First Order Reaction Plot

```python

import numpy as np
import matplotlib.pyplot as plt

#Plot for first order reaction [[Differential Method of Analysis]]

# -ln (CA)/(CAo)  v/s  t(seconds) with a slope of k
# Initial concentration of 1 (kmol/m^3)

t = np.linspace(0,180) #generating equally spaced points for time (seconds)
CA = np.linspace (1,0.7) #generating equally spaced points for concentration
CAo = 1 #(kmol/m^3)

# -ln (CA)/(CAo) has been represented by 'x'

x = -(np.log10(CA/CAo))
 

plt.plot(t,x)
plt.xlabel('t (sec)')
plt.ylabel('-ln (CA)/(CAo)')

plt.show()

#taking the question forward. Say we want to find out 'k' for first order 
#reaction

# A---> products (First Order Rxn)
# From the method of least squares, for fitting the curve of form y = mx:
# value of m can be obtained from m = [sum (yi * xi)/ sum (x^2)]
# Here m = k; y = -ln (CA)/(CAo) ; x = t

#prod_1 = yi * xi

prod_1 = x*t
prod_2 = t*t


k = sum (prod_1)/ sum (prod_2)  #method of least squares to find the rate constant
print(k) 


```
